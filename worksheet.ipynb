# Wine Quality Prediction

We use the wine quality dataset avalaible on Internet. This dataset has the fundamental features which are responsible for affecting the quality of the wine. By the use of several Machine learning models, we will predict the quality of the wine.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('winequalityN.csv')
print(df.head())
```

```python
df = df.drop('total sulfur dioxide', axis=1)
```

# Model Develppment

```python
df['best quality'] = [1 if x > 5 else 0 for x in df.quality]
```

```python
df.replace({'white' : 1, 'red' : 0}, inplace=True)
```

```python
# Mengisi nilai NaN dengan rata-rata
features = df.drop(['quality', 'best quality'], axis=1)  # Mendefinisikan ulang `features` dari awal
features = features.fillna(features.mean())

# Menentukan target
target = df['best quality']

# Membagi data menjadi train dan test
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(
    features, target, test_size=0.2, random_state=40
)

# Menggunakan SimpleImputer untuk imputasi nilai yang hilang
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')  # Menggunakan huruf kecil untuk `strategy`
xtrain = imputer.fit_transform(xtrain)
xtest = imputer.transform(xtest)

# Menampilkan bentuk data
xtrain.shape, xtest.shape
```

```python
norm = MinMaxScaler()
xtrain = norm.fit_transform(xtrain)
xtest = norm.transform(xtest)
```

```python
models = [LogisticRegression(), XGBClassifier(), SVC(kernel='rbf')]

for i in range(3):
    models[i].fit(xtrain, ytrain)

    print(f'{models[i]} : ')
    print('Training Accuracy : ', metrics.roc_auc_score(ytrain, models[i].predict(xtrain)))
    print('Validation Accuracy : ', metrics.roc_auc_score(
        ytest, models[i].predict(xtest)
    ))

    print()
```

# Model Evalution

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(ytest, models[1].predict(xtest))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=models[1].classes_)
disp.plot()
plt.show()
```

```python
print(metrics.classification_report(ytest, models[1].predict(xtest)))
```
